# Fill Out Ans

```python
import pandas as pd

ques_path = 'sample.csv'
ans_path = 'my_ans_tab_lr.csv'

q_df = pd.read_csv(ques_path)
data = pd.read_csv(ans_path)

```

```python
for index, row in q_df.iterrows():
    question = str(row['序號'])

    print(question)

    year = int(question[:4])
    month = int(question[4:6])
    day = int(question[6:8])
    hour = int(question[8:10])
    min = int(question[10:12])
    loc = int(question[12:14])

    result = data[(data['Month'] == month) & 
            (data['Day'] == day) & 
            (data['Hour'] == hour) & 
            (data['Minute'] == min) & 
            (data['Location'] == loc)]

    # 如果有匹配的结果，输出 Power
    if not result.empty:
        power_value = result['Power'].iloc[0]  # 获取第一个匹配的 Power 值
        print(f"The Power is: {power_value}")
        print(len(result['Power']))

        q_df.loc[index, '答案'] = power_value

    else:
        print("No matching data found.")

    #print(year, month, day, hour, min, loc)
q_df.to_csv('lr-v2.csv', index=False)
```
