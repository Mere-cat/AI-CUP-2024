# 整理成長表

```python
import os
import pandas as pd
import csv

columns=['year', 'month', 'day', 'hour', 'Precp']

data_files = [f for f in os.listdir() if f.endswith('.csv')]
data_files = sorted(data_files)
output_file = pd.DataFrame()

for month_file in data_files:
    print(month_file)
    df = pd.read_csv(month_file)
    year = month_file.split('-')[1]
    month = month_file.split('-')[2]

    output_row = pd.DataFrame([{col: None for col in columns}])
    for index, row in df.iterrows():
        if(row['日/時'] == '總和'):
            continue
        day = row['日/時']

        output_row['year'] = year
        output_row['month'] = month
        output_row['day'] = day

        for hour in range(1, 25):
            output_row['hour'] = hour
            output_row['Precp'] = row[hour]
            output_file = pd.concat([output_file, output_row], ignore_index=False)

    # 查看結果
    print(output_file)

output_file.to_csv('test/test.csv')
        
```

```python
df = pd.read_csv('test/test.csv')

minutes_list = [0, 10, 20 ,30, 40, 50] # 每小時分成 6 組，每組代表 10 分鐘
new_rows = []

# 將原始每行分裂成多行
for _, row in df.iterrows():
    for minutes in minutes_list:
        new_row = row.copy()
        new_row['minutes'] = minutes
        new_rows.append(new_row)

# 生成新的 DataFrame
expanded_df = pd.DataFrame(new_rows)
expanded_df.to_csv('test/test2.csv')

print(expanded_df)
```
