# 資料整理

```python
input_path = 'raw_data'
output_path = 'GlobalSolarRadiation'
```

```python
import pandas as pd
import numpy as np
import os
```

```python
# 先依時間順序排列
dir_list = os.listdir(input_path)
dir_list.sort()
```

```python
# 將 Precipitation 移到 Precipitation 目錄
for each_file in dir_list:
    if(each_file.find('Precipiation')):
        ori_path = os.path.join('raw_data', each_file)
        tar_path = os.path.join('Precipitation', each_file)
        os.rename(ori_path, tar_path)
```

```python


for each_file in dir_list:
    year = each_file[7:11]
    month = each_file[12:14]
    day = each_file[15:17]
    print(each_file, year, month, day)
```
